```python
import pandas as pd 
from porise.feature_engineering.preprocessing import prepare_df
```

```python
df = pd.read_csv("../porise/data/exp5217_all_imp100_merged.csv")
```

```python
df.head()
```

```python
cat_cols = ['first_point_rakuten_subservice_cd', 
            'first_earned_point_rakuten_subservice_cd', 
            'first_used_point_rakuten_subservice_cd', 
            'first_login_rakuten_subservice_cd', 
            'last_earned_point_rakuten_subservice_cd', 
            'ichiba_order_device_preference_cd', 
            'last_login_rakuten_subservice_cd', 
            'last_used_point_rakuten_subservice_cd', 
            'last_point_rakuten_subservice_cd',
            'reg_prefecture_cd']
num_cols = list(set(df.columns.tolist()) - set(cat_cols) - set(['easy_id']))
assert len(df.columns.tolist()) == len(num_cols) + len(cat_cols) + 1
```

```python
# for cat in cat_cols:
#     print(cat, df[cat].unique().tolist())
```

```python
cat_cols_remove = [cat for cat in cat_cols if len(df[cat].unique().tolist()) > 50]
cat_cols = list(set(cat_cols) - set(cat_cols_remove))
print(cat_cols)
```

```python
df = df.drop(columns=cat_cols_remove)
```

```python
df = prepare_df(df, num_cols=num_cols, cat_cols=cat_cols)
# print(df.columns.tolist())
```
